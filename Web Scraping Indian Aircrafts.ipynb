# Web Scraping Indian Military Aircrafts

```python
# Importing necessaey libraries
import requests
from bs4 import BeautifulSoup
```

```python
# Getting the response from the url
response = requests.get("https://en.wikipedia.org/wiki/List_of_active_Indian_military_aircraft")
```

```python
# Checking the response code
response
```

```python
# Getting the html content
html_content = response.content
```

```python
html_content
```

```python
# Turning the html code into more readable code
soup=BeautifulSoup(html_content,'html.parser')
```

```python
soup
```

```python
# Getting the required table from the html content
table = soup.find('table',class_='wikitable')
```

```python
table
```

```python
aircrafts = []

# Saving the aircraft names into aircraft variables
for aircraft in table.find_all('tr')[2:]:
    aircrafts.append(aircraft.find('a').text)
```

```python
aircrafts
```

```python
# importing csv module to to store state names into a csv file
import csv
```

```python
with open('aircrafts.csv', 'w', newline='',encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile) 
    writer.writerow(['Aircraft Name']) # Writing the header of the csv file
    for aircraft in aircrafts:
        print(aircraft)
        writer.writerow([aircraft]) # Writing Aircraft names into csv file
```
